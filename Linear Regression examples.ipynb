```python

def compute_model_output(x, w, b): # x is input variable, w is weight and b is bias
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      y (ndarray (m,)): target values
    """
    m = x.shape[0] # number of examples 
    f_wb = np.zeros(m) # it is used to store the predicted value of y for each x_i and it is initialized to zero
    for i in range(m): 
        f_wb[i] = w * x[i] + b # f_wb is the predicted value of y when x=x[i]
        
    return f_wb  
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,) # it will give the predicted value of y for each x_i in x_train

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction') # it will plot the line for predicted value of y for each x_i in x_train

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values') # it will plot the data points

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

**Now We compute cost Function for this Model**

```python
def compute_cost(x,y,w,b): # x is input variable, y is output variable, w is weight and b is bias
    m=x.shape[0] #number of training examples 
    cost_sum=0 # it is used to store the sum of cost for each training example and it is initialized to zero
    f_wb=0   # it is used to store the predicted value of y for each x_i and it is initialized to zero
    for i in range(m):

        f_wb=w*x[i]+b # f_wb is the predicted value of y 
        cost=(f_wb-y[i])**2 # this is cost function 
        cost_sum=cost_sum+cost # it will give the sum of cost for each training example
    total_cost=(1/(2*m))*cost_sum # this is the final cost function
    return total_cost 
```
