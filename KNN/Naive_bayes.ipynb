```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

**Data Preprocessing**

```python
from sklearn import datasets

iris = datasets.load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df=df.drop(df[df["target"]==2].index)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(df, df['target'], test_size=0.25,
                                                   random_state=1)
df=X_train
```

**Split the dataset**

```python
df1=df[df["target"]==0]
df2=df[df["target"]==1]
total=df.describe().iloc[0][4]
total1=df1.describe().iloc[0][4]
total2=df2.describe().iloc[0][4]
df.describe()
```

Figuring out the mean and standard deviation for each column

```python
f1=df1.describe()
f2=df2.describe()
m1=f1.iloc[1]
m2=f2.iloc[1]
std1=f1.iloc[2]
std2=f2.iloc[2]
```

Gaussian Function

```python
def gauss(a,b,c):
    d=-1*(((a-b)/c)**2)
    d=(math.exp(d))/c
    return d
```

```python
total=df.describe().iloc[0][4]
total1=df1.describe().iloc[0][4]
total2=df2.describe().iloc[0][4]
y_mtest=[]

for index,row in X_test.iterrows():
    p1=1
    p2=1
    for i in range(0,4):
        p1=p1*gauss(row[i],m1[i],std1[i])
    p1=p1*(total1/total)
    for i in range(0,4):
        p2=p2*gauss(row[i],m2[i],std2[i])

    p2=p2*(total2/total)

    if p1>p2:
        y_mtest.append(0)
    else:
        y_mtest.append(1)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_mtest ))
```

```python
from sklearn.naive_bayes import GaussianNB
mdl=GaussianNB()
X_train=X_train.drop("target",axis=1)
X_test=X_test.drop("target",axis=1)
mdl.fit(X_train,y_train)
y_pred=mdl.predict(X_test)
print(f"Accuracy of our model is : {accuracy_score(y_test,y_pred)*100}")
```
